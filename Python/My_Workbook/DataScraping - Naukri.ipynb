```python
import urllib.request
from bs4 import BeautifulSoup
import time
```

```python
data_list = []
def Naukri_WebScraper(url):
    source = urllib.request.urlopen(url).read()
    monster_scrap = BeautifulSoup(source,'lxml')
    
    for data in monster_scrap.find_all('div',{'type':'tuple'}):
        title_in = data.find_all('li')[0].text
        hiringOrg = data.find('span',{'itemprop':'hiringOrganization'}).text
#        skills = data.find_all('span')[2].text
        skills = data.find('span',{'class':'skill'}).text
        jobLoc = data.find('span',{'itemprop':'jobLocation'}).text
        expReq = data.find('span',{'itemprop':'experienceRequirements'}).text
#        datePosted = data.find('div',{'itemprop':'datePosted'}).text
        data_str = title_in+";"+hiringOrg+";"+skills+";"+jobLoc+";"+expReq
        data_list.append(title_in+";"+hiringOrg+";"+skills+";"+jobLoc+";"+expReq)
#    print(data_list[0])

#Naukri_WebScraper('https://www.naukri.com/data-scientist-jobs', data_list)
```
