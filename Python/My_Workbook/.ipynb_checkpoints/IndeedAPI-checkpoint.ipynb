```python
import urllib.request
import xml.etree.ElementTree as ET
import csv
import time
```

```python
def indeed_api(api):
    return(urllib.request.urlopen(api).read())
    
```

```python
def indeed_jobs_data(jobTitle = None, location = None, country = None, pageIterations = 1):
    startNumber = 0
    job_data_list = []
    api_key = '3864921654995829'
    counter = 1
    csv_header = ['Jobkey','Jobtitle','Company','City','State','Country','FormattedLocation','Date','Snippet','URL','Latitude','Longitude','FormattedRelativeTime']
    for startNumber in range(pageIterations):
        
        api_url = 'http://api.indeed.com/ads/apisearch?publisher='+str(api_key)+'&q='+str(jobTitle)+'&l='+str(location)+'&sort=&radius=150&st=&jt=&start='+str(startNumber)+'&limit=25&fromage=&filter=&latlong=1&co='+str(country)+'&chnl=&userip=1.2.3.4&useragent=Mozilla/%2F4.0%28Firefox%29&v=2'
        startNumber += 24
        print(api_url)
        job_data_xml = ET.fromstring(indeed_api(api_url))
        print('API called for page iteration '+str(counter))
        counter = counter + 1
        
        job_results = job_data_xml.findall('results/result')
        
        for result in job_results:
            data = []
            jobkey = result.find('jobkey').text
            jobtitle = result.find('jobtitle').text
            company = result.find('company').text
            city = result.find('city').text
            state = result.find('state').text
            country = result.find('country').text
            formattedLocation = result.find('formattedLocation').text
            date = result.find('date').text
            snippet = result.find('snippet').text
            url = result.find('url').text
            if result.find('latitude') is not None:
                latitude = result.find('latitude').text
                longitude = result.find('longitude').text
            else:
                latitude = None
                longitude = None
            formattedRelativeTime = result.find('formattedRelativeTime').text
            
            data.extend([jobkey, jobtitle,company, city, state, country, formattedLocation, date, snippet, url, latitude, longitude, formattedRelativeTime])
            job_data_list.append(data)
            
        time.sleep(3)
    
    print('Writing to csv file')
    with open(str(jobTitle)+'_'+str(location)+'.csv', 'w', encoding='utf-8', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(csv_header)
        writer.writerows(job_data_list)
```
