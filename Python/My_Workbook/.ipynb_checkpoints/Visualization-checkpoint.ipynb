```python
#Consolidated method to do all clean up and visualization

def job_vis(csv_file):
    df_skills = pd.read_csv(csv_file)

    print('Skills required for a data scientist are: ')
    display(df_skills['Skill'])
    
    print('Top 10 skills looked for by companies WRT job postings are: ')
    df_skills=df_skills.sort_values(['Count'], ascending=[False])
    display(df_skills.head(10))
    
    print("Let's see this data in bar graph")
    df_skills.head(20).plot(kind = 'bar', x='Skill', y='Count')
```

```python
job_vis('skills_data+scientist_india.csv')
```
