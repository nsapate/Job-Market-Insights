```python
data = driver.page_source
```

```python
soup = BeautifulSoup(data,'lxml')
soup
```

```python
data_to_write = []
for text in soup.find_all('li',{'class':"empReview padVert cf "}):
        heading = text.find('span',{'class':'summary'}).text
        heading = heading[1:-1]
        pros = text.find('p',{'class':'pros noMargVert truncateThis wrapToggleStr'}).text
        cons = text.find('p',{'class':'cons noMargVert truncateThis wrapToggleStr'}).text
        stars = len(text.find_all('i',{'class':'star'}))
        data_to_write.append(heading+";"+str(stars)+";"+pros+";"+cons)
        
data_to_write        
        
```
