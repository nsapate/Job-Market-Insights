```python
with open('monster.txt','w', encoding='utf-8') as file:
    for data in monster_scrap.find_all('div',{'class':'jobwrap'}):
        file.write(data.text)
```

```python
def Monster_WebScraper(url, data_list):
    source = urllib.request.urlopen(url).read()
    monster_scrap = BeautifulSoup(source,'lxml')
    
    for data in monster_scrap.find_all('div',{'class':'jobwrap'}):
        title_in = data.find('span',{'class':'title_in'}).text
        hiringOrg = data.find('span',{'itemprop':'hiringOrganization'}).text
        skills = data.find_all('span')[2].text
        jobLoc = data.find('span',{'itemprop':'jobLocation'}).text
        expReq = data.find('span',{'itemprop':'experienceRequirements'}).text
        datePosted = data.find('div',{'itemprop':'datePosted'}).text
        data_str = title_in+";"+hiringOrg+";"+skills+";"+jobLoc+";"+expReq+";"+datePosted
        data_list.append(data_str)
    with open('monsterScrapTest.txt','w', encoding='utf-8') as file:
        for data in data_list:
            file.write(data)
            file.write("\n")
        file.close()
        
        
```

```python
data_list =[]
Monster_WebScraper('http://www.monsterindia.com/data-scientist-jobs.html',data_list)

for index in range(2,501):
    Monster_WebScraper('http://www.monsterindia.com/data-scientist-jobs-'+index+'.html',)
```
